```python
import pandas as pd
```

```python
movies=pd.read_csv("/content/movies.dat",delimiter='::')
movies.columns=["ID", "Title", "Genre"]
movies.head()
```

```python
rating= pd.read_csv('/content/ratings.dat',delimiter='::')
rating.columns=["User", "ID", "Ratings", "Timestamp"]
rating.head()
```

```python
data=pd.merge(movies,rating,on=['ID','ID'])
data.head()
```

```python
ratings= data['Ratings'].value_counts()
numbers= ratings.index
quantity= ratings.values
import plotly.express as px
fig=px.pie(data,values=quantity, names=numbers)
fig.show()
```

```python
data2=data[['Title','Ratings']]
df=data2[data2['Ratings']==10]
df.groupby('Title').count().sort_values(by='Ratings',ascending=False).head(10)
```

```python
data2=data[['Title','Ratings']]
df=data2[data2['Ratings']==1]
df.groupby('Title').count().sort_values(by='Ratings',ascending=False)
```
